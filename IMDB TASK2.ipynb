```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\hp\Downloads\archive (1).zip",encoding='latin-1')
df.head()
```

```python
df = df.dropna()
df.head(10)
```

```python
df['Genre'] = df['Genre'].str.split(', ')
```

```python
df.head()
```

```python
mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(df['Genre']),columns=mlb.classes_)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
features = ['Director', 'Actor 1', 'Actor 2', 'Actor 3']
```

```python
onehot = OneHotEncoder(sparse=False, handle_unknown='ignore')
features_encoded = pd.DataFrame(onehot.fit_transform(df[features]),columns=onehot.get_feature_names_out(features))
```

```python
x = pd.concat([genre_encoded, features_encoded], axis=1)
y = df['Rating']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
genre = ["Comedy","Drama","Musical"]
director = 'Rahul Rawail'
actor1 = 'Bobby Deol'
actor2 = 'Aishwarya Rai Bachchan'
actor3 = 'Shammi Kapoor'
```

```python
genre_enc = mlb.transform([genre])
features_enc = onehot.transform([[director,actor1,actor2,actor3]])
```

```python
new_feature = np.concatenate([genre_enc, features_enc], axis =1)
```

```python
pred = model.predict(new_feature)
pred
```
