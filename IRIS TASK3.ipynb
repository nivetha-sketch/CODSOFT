```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species'],categories=pd.factorize(df['species'])
df.head()
```

```python
df.describe
```

```python
df.isna().sum()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df.petal_length,df.petal_width,df.species)
ax.set_xlabel('PETALLENGTHINCM')
ax.set_ylabel('PETALWIDTHINCM')
ax.set_zlabel('Species')
plt.title('3D SCATTER PLOT')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df.sepal_length,df.sepal_width,df.species)
ax.set_xlabel('SEPALLENGTHINCM')
ax.set_ylabel('SEPALWIDTHINCM')
ax.set_zlabel('Species')
plt.title('3D SCATTER PLOT')
plt.show()
```

```python
sns.scatterplot(data=df,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.scatterplot(data=df,x="petal_length",y="petal_width",hue="species")
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
   km=KMeans(n_clusters=k)
   km.fit(df[['petal_length','petal_width']])
   sse.append(km.inertia_)
```

```python
plt.xlabel('k_rng')
plt.ylabel('sum of squared errors')
plt.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3,random_state=0,)
y_predicted=km.fit_predict(df[['sepal_length','sepal_width']])
y_predicted
```

```python
df['clusters']=y_predicted
df.head(50)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(df.species,df.clusters)
cm
```

```python
true_labels=df.species
predicted_labels=df.clusters
cm=confusion_matrix(true_labels,predicted_labels)
class_labels=['setosa','versicolor','virginica']
plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues)
plt.title('Confusion matrix')
plt.colorbar()
tick_marks=np.arange(len(class_labels))
plt.xticks(tick_marks,class_labels)
plt.yticks(tick_marks,class_labels)
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j,i,str(cm[i][j]),ha='center',va='center',color='white')

plt.xlabel('predicted label')
plt.ylabel('True label')
plt.show()                
```
